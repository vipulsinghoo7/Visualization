# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

ANS--Seaborn is a popular data visualization library in Python that is built on top of Matplotlib. It provides a high-level interface for creating informative and visually appealing statistical graphics. Here are five commonly used plots in Seaborn and their uses:

1. Scatter Plot: A scatter plot displays the relationship between two numeric variables. It is useful for visualizing the correlation or patterns between two continuous variables. Seaborn provides additional functionality to enhance scatter plots, such as adding a regression line or color-coding the points based on a categorical variable.

2. Line Plot: A line plot shows the relationship between two variables with continuous data points connected by straight lines. It is commonly used to visualize trends over time or to compare multiple lines representing different categories. Seaborn offers options to customize the appearance of lines and markers, and it can also handle time-series data.

3. Bar Plot: A bar plot is used to compare categorical data by displaying rectangular bars with heights or lengths proportional to the values they represent. It is effective in showing the distribution or comparison of categorical variables. Seaborn allows customization of bar plots, such as adding error bars or grouping bars based on another categorical variable.

4. Histogram: A histogram is a graphical representation that organizes numerical data into bins and displays the frequency or count of values falling into each bin. It provides an overview of the data's distribution, allowing you to identify patterns, skewness, or outliers. Seaborn enhances histograms by providing options for kernel density estimation (KDE) plots or applying multiple histograms for different subgroups.

5. Heatmap: A heatmap represents data in a matrix format, where the values in the matrix are displayed using color variations. It is useful for visualizing the correlation or relationships between multiple variables simultaneously. Seaborn's heatmap offers customization options such as color palettes, annotations, and clustering, making it a powerful tool for exploring complex datasets.

These are just a few examples of the plots available in Seaborn. The library also provides other plots like "box plots", "violin plots", "count plots", and more, each with its specific use case for data analysis and visualization.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
import matplotlib.pyplot as plt
sns.lineplot(x = 'timepoint' ,  y = 'signal' , data = fmri , hue='event' , style='region')
plt.show()
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
titanic
plt.figure(figsize = (5,5))
plt.subplot(121)
sns.boxplot(data = titanic , x = 'pclass', y = 'age' )
plt.title("box plot of age vs pclass")
plt.subplot(122)
sns.boxplot(data = titanic , x = 'pclass', y = 'fare')
plt.title("box plot of fare vs pclass")
plt.tight_layout()
plt.show()
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
diamonds = sns.load_dataset("diamonds")
sns.histplot(data = diamonds , x = 'price' , hue ='cut' ,kde = True  )
plt.title("histogram of diamond price by cut")
plt.xlabel("price")
plt.ylabel("frequency")
plt.show()
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
sns.pairplot(data = iris , hue ="species" )
plt.title("pairplot of iris dataset")
plt.show()
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')
flights = flights.pivot('month','year','passengers')
sns.heatmap(data=flights , annot=True , fmt='d',cmap='YlGnBu')
plt.title("passengers flights")
plt.xlabel("year")
plt.ylabel("month")
plt.show()
```
